### Weekly bike usage

```python
def total_weekly_bike_usage(yyyy, w, dataset):
    
    year_data = dataset[dataset.year == yyyy]
    week_data = year_data[year_data.week == w]

    week_usage = week_data["usage"]
    print(sum(abs(week_usage)))
```
